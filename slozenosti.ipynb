```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from([2, 3,4,5,6,7])
```

```python
G.add_edges_from([(1, 2), (1, 3),(3,5),(5,1)])
```

```python

nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold')
```
